```python
total_revenue = clean_df["Revenue"].sum()
total_revenue
```

### Total Revenue
Total revenue represents the overall sales value generated during the observed period.
It provides a high-level view of business performance.


```python
monthly_revenue = (
    clean_df
    .set_index("InvoiceDate")
    .resample("ME")["Revenue"]
    .sum()
)

```

### Monthly Revenue Trend
Monthly revenue shows how business performance changes over time.
This helps identify growth patterns, seasonality, or early signs of decline.


```python
order_revenue = clean_df.groupby("Invoice")["Revenue"].sum()
aov = order_revenue.mean()
aov
```

### Average Order Value (AOV)
AOV measures the average revenue generated per order.
It helps evaluate pricing strategy and customer purchasing behavior.


```python
customer_orders = clean_df.groupby('Customer ID')["Invoice"].nunique()

repeat_customers = (customer_orders > 1).sum()
total_customers = customer_orders.count()

repeat_customer_rate = repeat_customers / total_customers
repeat_customer_rate
```

### Repeat Customer Rate
This metric measures the proportion of customers who made more than one purchase.
A higher repeat rate indicates stronger customer loyalty and retention.


```python
customer_revenue = clean_df.groupby('Customer ID')["Revenue"].sum()

top_10_percent_cutoff = int(0.1 * len(customer_revenue))

top_customers_revenue = (
    customer_revenue
    .sort_values(ascending=False)
    .head(top_10_percent_cutoff)
    .sum()
)

revenue_concentration = top_customers_revenue / customer_revenue.sum()
revenue_concentration
```

### Revenue Concentration (Top 10% Customers)
This metric measures how much revenue comes from the top 10% of customers.
High concentration indicates dependency risk on a small customer segment.


## Key Insights

- Revenue is unevenly distributed, with a small percentage of customers contributing a large share of total sales.
- Monthly revenue trends show noticeable variation, suggesting seasonality or demand fluctuations.
- Average order value indicates that most transactions are relatively small, with occasional high-value orders.
- A meaningful portion of customers place repeat orders, highlighting opportunities for retention strategies.


```python
import matplotlib.pyplot as plt

monthly_revenue.plot(title="Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
import matplotlib.pyplot as plt

clean_df["Revenue"].plot(kind="hist", bins=50, title="Revenue Distribution")
plt.xlabel("Revenue per Transaction")
plt.ylabel("Frequency")
plt.show()
```

**Observation:**  
Revenue distribution is highly right-skewed, indicating that most transactions are low value,
with a small number of very high-value purchases.


```python
customer_revenue = (
    clean_df.groupby('Customer ID')["Revenue"]
    .sum()
    .sort_values(ascending=False)
)

customer_revenue.head(50).plot(kind="bar", title="Revenue from Top 50 Customers")
plt.xlabel("Customer Rank")
plt.ylabel("Total Revenue")
plt.show()
```

**Observation:**  
Revenue contribution is heavily concentrated among top customers,
indicating dependency risk on a small customer segment.


```python
customer_orders = clean_df.groupby('Customer ID')["Invoice"].nunique()

repeat_counts = customer_orders.value_counts().sort_index()

repeat_counts.plot(kind="bar", title="Number of Orders per Customer")
plt.xlabel("Number of Orders")
plt.ylabel("Number of Customers")
plt.show()
```

**Observation:**  
A large proportion of customers place only one order,
while a smaller group contributes repeat purchases.
This highlights potential for retention-focused strategies.
